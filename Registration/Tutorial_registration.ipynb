```python
import shutil
import os

root_path = '/data/ceiling/workspace/HCC/Registration/Fine_Registration/'
save_if = '/data/ceiling/workspace/HCC/Registration/IF'
save_he = '/data/ceiling/workspace/HCC/Registration/HE'
for patient_name in os.listdir(root_path):
    patient_path = os.path.join(root_path, patient_name, 'save_image')
    print(patient_name)
    for name in os.listdir(patient_path):
        if name[0] == '.':
            continue
        source_path = os.path.join(patient_path, name)
        if 'Scan' in name:
            save_if_path = os.path.join(save_if, name)
            shutil.copyfile(source_path, save_if_path)
        else:
            save_he_path = os.path.join(save_he, name)
            shutil.copyfile(source_path, save_he_path)
```
